# Matplotlib Line Graph Example - Mechanical Engg - FEA 

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Locations along the bridge with temperature variation
Loc=['Loc1','Loc2','Loc3','Loc5']
#Designs with variation in temperature
D1=[400,450,500,570]
D2=[440,500,565,576]
#life cycle for corresponding designs
L1=[2300,2200,2000,1500]
L2=[2500,2500,1900,1000]
```

```python
# We shall plot location vs design vs Life
plt.xlabel("Failure Location")
plt.ylabel("Temerature (K)")
plt.plot(Loc,D1)
plt.plot(Loc,D2)
plt.plot(Loc,L1)
plt.plot(Loc,L2)
plt.title('Fire Deck Fatigue Failure',fontsize=20,fontweight=6,color='blue')
```

```python
# we see that both parameters (Temoerature and Life) cant be plotted wrt Location simultaneously, hence we require an additional y axis for plotting the data
#But befor that we shall add the legends, grids and data points to the plot


plt.xlabel("Failure Location")
plt.ylabel("Temerature (K)")

# Temperature is plotted with dashed line even other line style can be selected, color parameter can be used to set colours, and even marker can be selected
plt.plot(Loc,D1,'bo',linewidth=3)
#also we can use     plt.plot(Loc,D2,marker='>',color='green',linewidth=3,linestyle='dashed')
plt.plot(Loc,D2,'g>--',linewidth=3)
# Life we shall use different lines
plt.plot(Loc,L1,'m8-.',linewidth=3)
plt.plot(Loc,L2,'rv:',linewidth=3)
plt.title('Fire Deck Fatigue Failure',fontsize=20,fontweight=6,color='blue')
```

```python
# we can also visualize on the points 
plt.xlabel("Failure Location")
plt.ylabel("Temerature (K)")
plt.plot(Loc,D1,'bo',label="T1")
plt.plot(Loc,D2,'g>',label="T2")
plt.plot(Loc,L1,'m8',label="L1")
plt.plot(Loc,L2,'rv',label="L2")
plt.title('Fire Deck Fatigue Failure',fontsize=20,fontweight=6,color='blue')
#plotting the legend 
plt.legend(loc='best',shadow="True",fontsize="large")
#plot the grids
plt.grid(color='black', linestyle='--', linewidth=2,alpha=.1)
```

```python
#Now we shall additional axis to visulaize both temperature and life cycle 
#import the following libraries
# ref obtained from matplotlib documentation
from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist

host = host_subplot(111, axes_class=axisartist.Axes)
plt.subplots_adjust(right=0.75)

par1 = host.twinx()
#par2 = host.twinx()

#par2.axis["right"] = par2.new_fixed_axis(loc="right", offset=(60, 0))

par1.axis["right"].toggle(all=True)
#par2.axis["right"].toggle(all=True)




# setting the graph plot limits
host.set_xlim(-1, 5)
host.set_ylim(0, 3000)
par1.set_ylim(400,600)
#par2.set_ylim(1, 65)


host.set_xlabel("Failure Location")
host.set_ylabel("Fatigue Life (cycles)")
par1.set_ylabel("Temperature (K)")
plt.plot(Loc,D1,'mD',linewidth=0.5,label="Temp_Design1")
plt.plot(Loc,D2,'g>--',linewidth=0.5,label="Temp_Design2")
plt.plot(Loc,L1,'b8-.',linewidth=3,label="Life_Design1")
plt.plot(Loc,L2,'r:',linewidth=3,label="Life_Design2")
plt.title('Fire Deck Fatigue Failure',fontsize=20,fontweight=6,color='blue')
#plotting the legend 
plt.legend(loc='best',shadow="True",fontsize="large")
#plot the grids
plt.grid(color='black', linestyle='--', linewidth=2,alpha=.1)




#plotting the legend 
host.legend(loc='best',shadow="True",fontsize="large")
#plot the grids
host.grid(color='black', linestyle='--', linewidth=2,alpha=.1)


plt.show()
```
