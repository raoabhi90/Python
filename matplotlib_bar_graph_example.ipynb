```python
#import the matplotlib library for plotting the graph and numpy for numerical operations
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Life cycle failure data for Failure locations Loc 1 , Loc 14, Loc 5, Loc 8, Loc11

stress_loc = ['loc1','loc14','loc5','loc8','loc11']

#Let us take failures for 4 different designs where Life is influnecced by water jacket height and valve bridge thickness

failure1 = [2000,1400,8000,3000,6000]
failure2 = [2050,1600,7000,2000,7000]
failure3 = [2400,1700,7500,6000,8000]
failure4 = [2600,1200,6000,4000,5000]
```

```python
# Plotting Failure Location vs Life cycle (Bar graph) to check the design influence on Life cycle . To further investigate design parameter vs failure location  - life cycle heat map or cross validation in Machine learning can be explored 

plt.bar(stress_loc,failure1,label='Design1')
plt.bar(stress_loc,failure2,label='Design2')
plt.bar(stress_loc,failure3,label='Design3')
plt.bar(stress_loc,failure4,label='Design4')

#Plotting the Title

plt.title('Fatigue Failure',fontsize=20,fontweight=6,color='blue')

#Plot the x axis label
plt.xlabel("Failure Locations",fontsize=16,fontweight=4,color='brown')

#Plot the y axis label
plt.ylabel("Life Cycle",fontsize=16,fontweight=4,color='brown')

#plotting legends to identify the designs
plt.legend()
```

```python
#since its overlapping on same line its difficult to identify, hence we can offset the bins and control its width for which we shall use numpy to find the lenght and map it with locatino using Xtick option 

xpos=np.arange(len(stress_loc))
xpos
```

```python
# xpos- distance to get the clear view of bars

plt.bar(xpos-0.1,failure1,width=0.1,label='Design1')
plt.bar(xpos-0.2,failure2,width=0.1,label='Design2')
plt.bar(xpos-0.3,failure3,width=0.1,label='Design3')
plt.bar(xpos-0.4,failure4,width=0.1,label='Design4')
plt.xticks(xpos,stress_loc)

#Plotting the Title

plt.title('Fatigue Failure',fontsize=20,fontweight=6,color='blue')

#Plot the x axis label
plt.xlabel("Failure Locations",fontsize=16,fontweight=4,color='brown')

#Plot the y axis label
plt.ylabel("Life Cycle",fontsize=16,fontweight=4,color='brown')

#plotting legends to identify the designs  loc =4 indicates the legend postion in the plot refer documentation of matplot lib
plt.legend(loc=2,shadow="True",fontsize="large")

#plotting the grid, alpha represents the transparency which varies from 0 to 1 
plt.grid(color='black', linestyle='--', linewidth=2,alpha=.3)



```

```python
############################################################
#limits of graph can be set 
from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist


host = host_subplot(111, axes_class=axisartist.Axes)
plt.subplots_adjust(right=0.75)

# limit can be set for axis
host.set_xlim(-1, 4)
host.set_ylim(0, 9000)




############################################################
# xpos- distance to get the clear view of bars

plt.bar(xpos-0.1,failure1,width=0.1,label='Design1')
plt.bar(xpos-0.2,failure2,width=0.1,label='Design2')
plt.bar(xpos-0.3,failure3,width=0.1,label='Design3')
plt.bar(xpos-0.4,failure4,width=0.1,label='Design4')
plt.xticks(xpos,stress_loc)

#Plotting the Title

plt.title('Fatigue Failure',fontsize=20,fontweight=6,color='blue')

#Plot the x axis label
plt.xlabel("Failure Locations",fontsize=16,fontweight=4,color='brown')

#Plot the y axis label
plt.ylabel("Life Cycle",fontsize=16,fontweight=4,color='brown')

#plotting legends to identify the designs  loc =4 indicates the legend postion in the plot refer documentation of matplot lib
plt.legend(loc=2,shadow="True",fontsize="large")

#plotting the grid, alpha represents the transparency which varies from 0 to 1 
plt.grid(color='black', linestyle='--', linewidth=2,alpha=.3)




```

```python
#Graph can be converted to horizontal axis by adding plot.barh, width change to height and xticks to yticks


############################################################
# xpos- distance to get the clear view of bars

plt.barh(xpos-0.1,failure1,height=0.1,label='Design1')
plt.barh(xpos-0.2,failure2,height=0.1,label='Design2')
plt.barh(xpos-0.3,failure3,height=0.1,label='Design3')
plt.barh(xpos-0.4,failure4,height=0.1,label='Design4')
plt.yticks(xpos,stress_loc)

#Plotting the Title

plt.title('Fatigue Failure',fontsize=20,fontweight=6,color='blue')

#Plot the x axis label
plt.ylabel("Failure Locations",fontsize=16,fontweight=4,color='brown')

#Plot the y axis label
plt.xlabel("Life Cycle",fontsize=16,fontweight=4,color='brown')

#plotting legends to identify the designs  loc =4 indicates the legend postion in the plot refer documentation of matplot lib
plt.legend(loc=4,shadow="True",fontsize="large")

#plotting the grid, alpha represents the transparency which varies from 0 to 1 
plt.grid(color='black', linestyle='--', linewidth=2,alpha=.3)




```
